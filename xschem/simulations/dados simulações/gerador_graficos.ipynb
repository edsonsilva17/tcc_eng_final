```python
import csv
import matplotlib.pyplot as plt
x1 = list()
y1 = list()
x2 = list()
y2 = list()

#Abrindo dados PMOS 1.8 LVT
with open('diodo18lvt.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x1.append(float(linha[0]))
        y1.append(float(linha[1]))

#Abrindo dados PMOS 1.8
with open('diodo18.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x2.append(float(linha[0]))
        y2.append(float(linha[1]))

#Plotando Gráficos
plt.plot(x1,y1, lw=2., label='PMOS 1.8 LVT')
plt.plot(x2,y2, lw=2., label='PMOS 1.8')
#plt.ylim(0, 10)
plt.xlim(-2, 2)
plt.title('Curva caracteística - Mosfet funcionando como Diodo')
plt.xlabel('Tensão (V)')
plt.ylabel('Corrente (A)')
plt.legend()
plt.grid()
plt.savefig('curva_diodo.png', format='png')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
x1 = list()
y1 = list()
x2 = list()
y2 = list()
x3 = list()
y3 = list()

#Abrindo dados Entrada Comparador
with open('ota_in.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x1.append(float(linha[0]))
        y1.append(float(linha[1]))

#Abrindo dados Saída Comparador
with open('ota_out.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x2.append(float(linha[0]))
        y2.append(float(linha[1]))
        
#Abrindo dados Referência do Comparador
with open('ota_ref.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x3.append(float(linha[0]))
        y3.append(float(linha[1]))

#Plotando Gráficos
plt.plot(x1,y1, lw=2., label='Tensão de Entrada')
plt.plot(x2,y2, lw=2., label='Tensão de Saída')
plt.plot(x3,y3, lw=2., label='Tensão de Referência')
plt.xlim(0, 2)
plt.title('OTA como Comparador')
plt.xlabel('Tensão de Entrada (V)')
plt.ylabel('Tensão (V)')
plt.legend()
plt.grid()
plt.savefig('comparador.png', format='png')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

x1 = list()
y1 = list()
x2 = list()
y2 = list()
x3 = list()
y3 = list()
#Abrindo dados ret 1
with open('ret1.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x1.append(float(linha[0])*(10**6))
        y1.append(float(linha[1]))

#Abrindo dados ret 3
with open('ret3.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x2.append(float(linha[0])*(10**6))
        y2.append(float(linha[1]))

#Abrindo dados ret 8
with open('ret8.csv', 'r') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter=' ')
    for linha in arquivo_csv:
        while '' in linha:
            linha.remove('')
        x3.append(float(linha[0])*(10**6))
        y3.append(float(linha[1]))

#Plotando Gráficos
fig, axs = plt.subplots(1, 3, figsize=(16,4), dpi=300)
axs[0].plot(x1, y1, 'tab:red')
axs[0].set_title('1 Célula')
axs[1].plot(x2, y2, 'tab:blue')
axs[1].set_title('3 Células')
axs[2].plot(x3, y3, 'tab:green')
axs[2].set_title('8 Células')


for ax in axs.flat:
    ax.set(xlabel='Tempo (\u00B5s)', ylabel='Tensão (V)')
    ax.grid()


#for ax in axs.flat:
#    ax.label_outer()

plt.savefig('ret.png', format='png')
plt.show()
```
